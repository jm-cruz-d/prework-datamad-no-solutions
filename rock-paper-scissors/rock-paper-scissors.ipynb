```python
import random
lst = ['stone', 'paper', 'scissors']
games = 3
win = 2

def machine():
    return random.choice(lst)

def tu_turno():
    x = input("Enter 'stone', 'paper', 'scissors': ")
    if (x == 'stone') or (x == 'paper') or (x == 'scissors'):
        return x
    else:
        return input("Try again: ")


def combat():
    if machine() == tu_turno():
        return 0
    elif machine() == 'stone' and tu_turno() == 'scissors':
        return 1
    elif machine() == 'scissors' and tu_turno() == 'paper':
        return 1
    elif machine() == 'paper' and tu_turno() == 'stone':
        return 1
    else:
        return 2

def partida():
    suma1 = 0
    suma2 = 0
    if combat() == 1:
        suma1 +=1
    elif combat() == 2:
        suma2 +=1
    else:
        print("Tie")
    return (machine, tu_turno, "El juego va {} {}".format(suma1,suma2))

machine_wins = 0
human_wins = 0

while (machine_wins < win) or (human_wins < win):    
    z = combat()
    if z == 1:
        machine_wins += 1
    elif z == 2:
        human_wins += 1
    else:
        print("Tie")
    print(machine_wins, human_wins)

if machine_wins == win:
    print("Machine wins!!")
elif human_wins == win:
    print("Human wins!!")

#No se que me falla que solo me imprime cuando gana al que pongo en el if del Ãºltimo condicional y lo del elif no me lo coge
```
